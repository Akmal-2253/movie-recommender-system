```python
import ast


def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:      
            break
    return L    
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
movies['crew' ].iloc[0]
```

```python
import ast

def fetch_director(text):
    try:
        if isinstance(text, str):
            text = ast.literal_eval(text)
        
        for person in text:
            if person.get('job') == 'Director':
                return [person['name']]  # ← wrap in a list
        return []
    
    except (ValueError, TypeError):
        return []

```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(3)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x] if x is not None else [])
```

```python
movies.head(2)
```

```python
movies['tags']=movies ['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head(1)
```

```python
movies=movies.drop(columns=['overview','genres','keywords','cast','crew'],inplace=False)
movies.head(1)
```

```python
movies['tags']=movies['tags'].apply(lambda x: " ".join(x).lower())
```

```python
movies.head(3)
```

```python
movies['tags'][0]
```

```python
movies['tags'][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(movies['tags']).toarray() # type: ignore
vectors.shape
```

```python
import numpy as np
# forcing full output in vscode
np.set_printoptions(threshold=np.inf) # type: ignore

# Now run again
cv.get_feature_names_out()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem("'in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron'")
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
list(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
print(similarity.shape)
```

```python
print(type(similarity)) 
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1]) #with enumerate the index and the value of the similarity is given in a list of tuples and index is held in the first element and the value of similarity is held in the second element of the tuple
```

```python
def recommend(movie):
    movie_index=movies[movies['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
